```python
def count_stars(x, y, kernel_density=False, direction='r', bins=32, mask=None, bw_method=None):
    #compute density along the stream
    if mask=='None':
        mask=np.ones_like(x).astype(bool)
        
    rvals=(x**2+y**2)**0.5
    
    if direction=='x':
        rvals=x
        
    if direction=='y':
        rvals=y
     
    counts0, binedges= np.histogram(rvals, bins=bins, density=True)
    counts0=np.append(counts0,counts0[-1])
    assert len(binedges)== len(counts0)
    counts= np.interp(rvals, binedges ,counts0)
    assert len(counts)==len(rvals)
    
    if  bw_method is None:
        bw_method=np.nanstd(rvals)/1.1
        
    if kernel_density:
        kernel= stats.gaussian_kde(rvals, bw_method= bw_method)
        counts = kernel.pdf(rvals)
    return counts, rvals.flatten()
```

```python
def detect_overdens(dens, phi1, phi2, threshold=0.0, mask=None, verbose=False):
    if mask is None: mask=np.ones_like(phi1, phi2).astype(bool)
    x, y, density=phi1[mask], phi2[mask], dens[mask]
    peak_idxs=[]
    peaks,  _ = find_peaks(density, threshold=threshold)
    return peak_idxs
```

```python
def circular_mask(x, y, cent, radius):
    """
    Create a fake circular grap in a stellar stream
    
    
    Input: center, 3d-radius all in kpc
    
    returns a mask array
    """
    #compute absolute distance 
    r= ((x-cent[0])**2+(y-cent[1])**2)**0.5
    return  r> radius
```

```python
def custom_peak_finder(denst, npeaks, mask=None, step=10, threshold=0.,differential=False):
    """
    if differntial is true, it
    takes the gradient and returns the peaks in the first derivative
    
    """
    dens=np.copy(denst)
    
    if mask is None:
        mask=np.ones_like(dens).astype(bool)
    #fill nans above threshold and mask
    dens[~mask]=threshold/10

    if not differential:
        mask_total = np.logical_or(mask, dens>threshold)
        dens[~mask]=threshold/10
        #sort dens
        #pick peaks that are not close
        sep=int(step)
        peaks=(np.flip(np.argsort(dens))[::sep])[:npeaks]

        return peaks
    if  differential:
        delta_dens=np.diff(dens,  prepend=dens[0])
        peaks,_= find_peaks( delta_dens, threshold= threshold)
        return peaks[:npeaks]
```

```python
def detect_gap(x, y, step=5, broad_band=1., narro_band=0.05, mask=None, plot=False, npeaks=5,
              look_along='x', threshold=0., save=False, filename=None):
    """
    Gap detection function
    
    Input:
        x: x-axis (arbitray units)
        y: y-axis 
        step(optional): spacing in array step size to skip
        only reports peaks separated by at least step size 
        broad_band: the kde band for the overall density in stream
        narro_band: the kde band for variations in density in stream
        
    """
    #density, rvals=count_stars(streamgp, kernel_density=False, direction='r', bins=300)
    kdedensity, rvals=count_stars(x, y, kernel_density=True, direction= look_along, bw_method= narro_band)
    smooth_kdedensity,_ =count_stars(x, y, kernel_density=True, direction= look_along, bw_method= broad_band)
    
    #order by r and take 1/density
    sorted_idx=np.argsort(rvals)
    sorted_rs=rvals[sorted_idx]
    dens=abs(smooth_kdedensity/kdedensity)
    sorted_dens=dens[sorted_idx]
    if mask is None: mask= np.ones_like(x).astype(bool)
    sorted_mask=mask[sorted_idx]
    
    peaks=custom_peak_finder(sorted_dens,npeaks, mask= sorted_mask, threshold= threshold,  step=step)
    
    if plot:
        fig,ax=plt.subplots(figsize=(12, 4))
        ax.scatter(x, y, c='k', s=1, alpha=0.4)
        ax.plot(sorted_rs, sorted_dens, label='Scaled Density')
        ax.scatter(sorted_rs[peaks], np.ones_like(sorted_rs[peaks]), s=100, marker='x', c='r')
        ax.set_xlabel(r'$\phi_1$ (deg)', fontsize=16)
        ax.set_ylabel(r'$\phi_2$ (deg)', fontsize=16)
        ax.minorticks_on()
        plt.legend(fontsize=16)
        if save : plt.savefig(filename)

    return peaks
```

```python
#add backgorund and see if can detect can be 
def generate_uniform_background(n):
    Nstars = n *1000  #make these with 1,2,3,4...,10 times the amount of stars such that f = nstar/nbkg are the same.
    a = 0.3
    r = np.random.uniform(low=0, high=(a)**2, size = (Nstars)) #radius of scramble data, high=radius data **2         \
    angle = np.random.uniform(low=0, high=2.*np.pi, size = (Nstars ))
    x = np.sqrt(r) * np.cos(angle)
    y = np.sqrt(r) * np.sin(angle)
    return x, y
```

```python
#add background
rl_back_x, rl_back_y= generate_uniform_background(2)
rl_back_x, rl_back_y=rl_back_x*3.5, rl_back_y*3.5
rl_back_x=rl_back_x*np.ptp(rl_x)
rl_back_y=rl_back_y*np.ptp(rl_y)
#concatenate arrays 
rl_comb_x, rl_comb_y= np.concatenate([rl_x , rl_back_x+rl_x.mean()] ), \
np.concatenate([rl_y, rl_back_y+rl_y.mean()] )
```

```python
mask= np.logical_and( rl_comb_x <20,  rl_comb_x> 0.)
```
