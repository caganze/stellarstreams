```python
#coords standards
_ = astro_coord.galactocentric_frame_defaults.set('v4.0')

#potential and hamiltonian
pot=gp.MilkyWayPotential()
H = gp.Hamiltonian(pot)

#galactocentric reference frame
v_sun = astro_coord.CartesianDifferential([11.1, 220 + 24.0, 7.25]*u.km/u.s)

galcen_frame =astro_coord.Galactocentric(galcen_distance=8.*u.kpc,
                                    galcen_v_sun=v_sun)
```

```python

def read_stream_file(filename):
    """
    reads in a saved stream run 
    returns the subhalo, progenitor and stream
    """
    print (filename)
    f = h5py.File(path_data+'/'+filename, mode='r')
    
    body=f['nbody']
    stream_cont=f['stream']
    
    #f.close()
    return body, stream_cont

def read_orbit_file(filename):
    return np.load(path_data+'/{}'.format(filename), allow_pickle=True)

def visualize_stream(st,ax, xlim=None, ylim=None, alpha=0.1, marker='+', s=1, color='k',\
                     yshift=0, plot_yz=False, xmask=None, ):
    if xmask is None:
        xmask=np.ones_like(st.x.value).astype(bool)
    if not plot_yz:
        ax.scatter(st.x.value[mask], st.y.value[mask]+yshift,  s=s, alpha=alpha,  marker=marker, color=color)
        ax.set(xlabel='x (kpc)', ylabel='z (kpc)')
    if plot_yz:
        ax.scatter(st.y.value, st.z.value+yshift,  s=s, alpha=alpha,  marker=marker, color=color)
        ax.set(xlabel='y (kpc)', ylabel='z (kpc)')
        
    if xlim is not None:
        ax.set(xlim=xlim)
    if ylim is not None:
        ax.set(ylim=ylim)
    return 

def evolve_stream_in_time(st, time, dt=1*u.Myr, progenitor=False):
    time_dict={'t1': 0*u.Myr, 't2': time, 'dt': dt}
    #make a phase space position objec
    stpos=None
    if progenitor:
        stpos=gd.PhaseSpacePosition(pos=np.array([st.x.value, st.y.value, st.z.value])*st.x.unit,
                              vel=np.array([st.v_x.value, st.v_y.value, st.v_z.value])*st.v_z.unit)
    if not progenitor:
        stpos=gd.PhaseSpacePosition(pos=np.array([st.x, st.y, st.z])*st.x.unit,
                              vel=np.array([st.v_x, st.v_y, st.v_z])*st.v_z.unit)
    final_stream_orbit=gp.Hamiltonian(pot).integrate_orbit(stpos,  **time_dict)
    
    #integrate in time
    new_stream_coord=SkyCoord(x=final_stream_orbit.x[-1], y=final_stream_orbit.y[-1],
                       z= final_stream_orbit.z[-1], v_x=final_stream_orbit.v_x[-1],\
                       v_y=final_stream_orbit.v_y[-1], v_z=final_stream_orbit.v_z[-1],\
                       frame=galcen_frame)
    return new_stream_coord
```

```python
filenames=np.array(glob.glob(path_data+'/*.npy'))
masses=np.array([float(x.split('mhalo')[-1].split('_')[0]) for x in filenames])
sorted_idx=np.argsort(masses)
masses_sort=masses[sorted_idx]
files_sort=filenames[sorted_idx]
```

```python
fig, ( ax,   ax1, ax2)=plt.subplots(figsize=(12, 8), ncols=3)
for idx in range(len(sts)):
    mask=np.logical_and.reduce([sts[idx].x.value > 9., sts[idx].y.value>0.0,
                                sts[idx].y.value<8.,
                              sts[idx].x.value < 10])
    mask1=np.logical_and(sts[idx].y.value > .5, sts[idx].y.value<1.5)
    #only sho an arm
    ax1.scatter(sts[idx].x.value[mask], sts[idx].y.value[mask]+idx,  s=1,\
               alpha=.2,  marker='*', color='#111111')
    

    ax.scatter(sts[idx].y.value[mask1], sts[idx].z.value[mask1]+idx,  s=1,\
               alpha=.2,  marker='*', color='#111111')
    
 
    ax2.scatter(sts[idx].y.value, sts[idx].z.value+idx,  s=1,\
               alpha=0.2,  marker='*', color='#111111')
    
    ax.text(0.5, np.nanmedian(sts[idx].z.value[mask]+idx)-0. , \
            " M = {:.0e} ".format(masses_sort[idx]) ,c='b', fontsize=15)



    ax.minorticks_on()

    ax.set(xlabel='y (pc)',  ylabel='z (kpc)')
    ax1.set(xlabel='x (pc)',  ylabel='y (kpc)')
    ax2.set(xlabel='y (pc)',  ylabel='z (kpc)')
    plt.tight_layout()
plt.savefig(path_plot+'/comparison_pannels_1Gyr.jpeg')
```

```python
from scipy import stats
```

```python
def make_density_plot(values, ax):
    kernel = stats.gaussian_kde(values)
    x, y=np.meshgrid(np.linspace(values[0].min(), values[0].max(), 60), \
                     np.linspace(values[1].min(), values[1].max(), 60))
    positions = np.vstack([x.ravel(), y.ravel()])
    Z = np.reshape(kernel(positions).T, x.shape)
    #ax.imshow(np.rot90(Z), cmap=plt.cm.gist_earth_r)
    ax.scatter(positions[0], positions[1], c=kernel(positions), cmap=plt.cm.gist_earth_r)
```

```python
#make kernel density
fig, (ax, ax1, ax2)=plt.subplots(figsize=(12, 8), ncols=3)
for idx in range(len(sts)):
    #only sho an arm
    off=idx+2*idx
    mask1=np.logical_and.reduce([evolved_sts[idx].x.value > -7, \
                        evolved_sts[idx].x.value<-1.,
                       evolved_sts[idx].y.value>0.
                                ])
    mask=np.logical_and.reduce([evolved_sts[idx].y.value > 2,\
                                 evolved_sts[idx].y.value<9])
    
    make_density_plot(np.array([evolved_sts[idx].y.value, evolved_sts[idx].z.value+off]), ax2)
    
    make_density_plot(np.array([evolved_sts[idx].y.value[mask],\
               evolved_sts[idx].z.value[mask]+off]), ax)
     
    make_density_plot(np.array([evolved_sts[idx].x.value[mask1],\
                                evolved_sts[idx].y.value[mask1]+off+5*idx]), ax1)
    
    ax.minorticks_on()
    ax2.set(xlim=[0, 10])
    
    ax.text(0.5, np.nanmedian(sts[idx].z.value[mask]+off-0.5) , \
            " M = {:.0e} ".format(masses_sort[idx]) ,c='b', fontsize=15)

ax.set(xlabel='y (pc)',  ylabel='z (kpc)')
ax1.set(xlabel='x (pc)',  ylabel='y (kpc)')
ax2.set(xlabel='y (pc)',  ylabel='z (kpc)')
plt.tight_layout()
plt.savefig(path_plot+'/comparison_pannels_1.5Gyr_density.jpeg')
```
