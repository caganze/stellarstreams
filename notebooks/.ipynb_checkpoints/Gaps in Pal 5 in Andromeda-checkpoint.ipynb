```python
#gap detector algorithm/functions
def count_stars(x, y, kernel_density=False, direction='r', mask=None, bw=None):
    #compute density along the stream
    if mask=='None':
        mask=np.ones_like(x).astype(bool)
        
    rvals=(x**2+y**2)**0.5
    
    if direction=='x':
        rvals=x
        
    if direction=='y':
        rvals=y
    
    counts=None
    if not kernel_density:
        counts0, binedges= np.histogram(rvals, bins=bw, density=True)
        counts0=np.append(counts0,counts0[-1])
        assert len(binedges)== len(counts0)
        counts= np.interp(rvals, binedges ,counts0)
        assert len(counts)==len(rvals)

    if  bw is None:
        bw =np.nanstd(rvals)/1.1
        
    if kernel_density:
        kernel= stats.gaussian_kde(rvals, bw_method= bw)
        counts = kernel.pdf(rvals)
    return counts, rvals.flatten()

def custom_peak_finder(denst, npeaks, mask=None, step=10, threshold=0.,differential=False):
    """
    if differntial is true, it
    takes the gradient and returns the peaks in the first derivative
    
    """
    dens=np.copy(denst)
    
    if mask is None:
        mask=np.ones_like(dens).astype(bool)
    #fill nans above threshold and mask
    dens[~mask]=threshold/10

    if not differential:
        mask_total = np.logical_or(mask, dens>threshold)
        dens[~mask]=threshold/10
        #sort dens
        #pick peaks that are not close
        sep=int(step)
        peaks=(np.flip(np.argsort(dens))[::sep])[:npeaks]

        return peaks
    if  differential:
        delta_dens=np.diff(dens,  prepend=dens[0])
        peaks,_= find_peaks( delta_dens, threshold= threshold)
        return peaks[:npeaks]
    
def detect_gap(x, y, step=5, broad_band=1., narro_band=0.05, mask=None, plot=False, npeaks=5,
              look_along='x', threshold=0., save=False, filename=None, ax=None, scale_dens_by=0.1, \
               use_kde=False, scatter_kargs={}, density_ax=None):
    """
    Gap detection function
    
    Input:
        x: x-axis (arbitray units)
        y: y-axis 
        step(optional): spacing in array step size to skip
        only reports peaks separated by at least step size 
        broad_band: the kde band for the overall density in stream
        narro_band: the kde band for variations in density in stream
        
    """
    #density, rvals=count_stars(streamgp, kernel_density=False, direction='r', bins=300)
    kdedensity, rvals=count_stars(x, y, kernel_density= use_kde, direction= look_along, bw= narro_band)
    smooth_kdedensity,_ =count_stars(x, y, kernel_density= use_kde, direction= look_along, bw= broad_band)
    
    #order by r and take 1/density
    sorted_idx=np.argsort(rvals)
    sorted_rs=rvals[sorted_idx]
    dens=abs(smooth_kdedensity/kdedensity)
    sorted_dens=dens[sorted_idx]
    if mask is None: mask= np.ones_like(x).astype(bool)
    sorted_mask=mask[sorted_idx]
    
    peaks=custom_peak_finder(sorted_dens,npeaks, mask= sorted_mask, threshold= threshold,  step=step)
    
    if plot:
        if ax is None : fig, ax=plt.subplots()
        ax.scatter(x, y, **scatter_kargs)
        if density_ax is None:  density_ax=plt.gca()
            
        density_ax.step(sorted_rs[ sorted_mask], \
                        (sorted_dens)[ sorted_mask], where='mid')
        
        #density_ax.step(rvals[sorted_idx][ sorted_mask], \
        #                np.nanmean(sorted_dens)*smooth_kdedensity[sorted_idx][ sorted_mask])
        
        #density_ax.step(rvals[sorted_idx][ sorted_mask],\
        #                np.nanmean(sorted_dens)*kdedensity[sorted_idx][ sorted_mask])
        
        density_ax.scatter(sorted_rs[peaks],\
                           np.ones_like(sorted_rs[peaks])*np.nanmedian(sorted_dens), \
                   s=300, marker='+', c='#0074D9')
        ax.scatter(sorted_rs[peaks], np.ones_like(sorted_rs[peaks])*np.nanmean(y[mask]),  \
                   s=300, marker='+', c='#0074D9')
        #ax.set_xlabel(r'$\phi_1$ (deg)', fontsize=16)
        #ax.set_ylabel(r'$\phi_2$ (deg)', fontsize=16)
        ax.minorticks_on()
        #.legend(fontsize=16)
        if save : plt.savefig(filename)

    return peaks

def rotate(x, y, ang):
    """
    Angle must be in radians
    """
    
    #rotation matrix
    r=[[np.cos(ang), -np.sin(ang)],
       [np.sin(ang), np.cos(ang)]]
    
    c=(np.nanmedian(x), np.nanmedian(y))
    
    i=np.identity(2)
    
    mat=np.matrix([[r[0][0], r[0][1], np.dot(i-r, c)[0]],
                   [r[1][0], r[1][1], np.dot(i-r, c)[1]],
                   [0., 0., 1.]])

    z=np.ones_like(x)
    
    rotated=np.array(np.dot(mat, np.array([x, y, z])))
    
    return rotated[0], rotated[1] 
```

```python
selected_1hr=np.array(data_1hr['stream'])
selected_bck_1hr=np.array(data_1hr['bckgd'])

selected_1000s=np.array(data_1000s['stream'])
selected_bck_1000s=np.array(data_1000s['bckgd'])
```

```python
scatter_kargs={'alpha':1., 's': 1., 'marker':'.', 'c':'#FFFFFF'}
fig, (ax1, ax)=plt.subplots(figsize=(10, 5), ncols=2, nrows=2, \
                            gridspec_kw={'height_ratios':[1, 3]}, sharex='col')

_= detect_gap(selected_1hr[0], selected_1hr[1], step=15, broad_band=1/int(2), narro_band=1/int(100),  npeaks=20,
              look_along='x',  ax=ax[0], plot=True, \
              use_kde=True, scatter_kargs=scatter_kargs, density_ax=ax1[0])
_= detect_gap(selected_1000s[0], selected_1000s[1], step=15, broad_band=1/int(2), narro_band=1/int(100),  npeaks=20,
              look_along='x',  ax=ax[-1],  plot=True, use_kde=True,
               scatter_kargs=scatter_kargs, density_ax=ax1[-1])
ax[0].set_title('Exposure: 1 hour ', fontsize=18,  y=-0.07)
ax[-1].set_title('Exposure: 1000 sec ', fontsize=18,  y=-0.07)
plt.tight_layout()
for a in ax:
    a.axis('off')
for a in ax1:
    #a.set_xlabel('x', fontsize=14)
    a.set_ylabel('Density Ratio' , fontsize=14)
plt.tight_layout()
plt.savefig(path_plot+'pal5_in_andromeda_roman_gaps.jpeg')
```
