```python
#imports
import astropy.units as u
import numpy as np
import gala.potential as gp
import gala.dynamics as gd
from gala.dynamics import mockstream as ms
from gala.units import galactic
from astropy.coordinates import SkyCoord, Distance
#import popsims
import matplotlib.pyplot as plt
#%matplotlib notebook
from tqdm import tqdm
import HSS
```

```python
#use NFW POTENTIAL BECAUSE THE HALO? 
pot = gp.NFWPotential.from_circular_velocity(v_c=220*u.km/u.s, r_s=15*u.kpc, units=galactic)
#CONSTRUCT A HAMILTONIAN
H = gp.Hamiltonian(pot)
```

```python
def generate_stream(mass, prog_w0, dt=8 * u.Myr, nsteps=1000):
    df = ms.FardalStreamDF()
    prog_mass = mass * u.Msun
    gen = ms.MockStreamGenerator(df, H)
    stream, prog = gen.run(prog_w0, prog_mass ,dt=dt, n_steps=nsteps)
    return stream, prog

def generate_uniform_background(n):
    Nstars = int(n *1000)  #make these with 1,2,3,4...,10 times the amount of stars such that f = nstar/nbkg are the same.
    a = 0.3
    r = np.random.uniform(low=0, high=(a)**2, size = (Nstars)) #radius of scramble data, high=radius data **2         \
    angle = np.random.uniform(low=0, high=2.*np.pi, size = (Nstars ))
    x = np.sqrt(r) * np.cos(angle)
    y = np.sqrt(r) * np.sin(angle)
    return x, y

def plot_hss(pos, HSS_args):
    HSS_args['pos']=pos[0], pos[-1]  #in deg here, can also be unitless
    HSS_args['kpc_conversion'] = np.pi * HSS_args['d_galaxy'] / 180.


    HSS.RT_plot(HSS_args['pos'], HSS_args['unit'], HSS_args['kpc_conversion'],\
                    HSS_args['delta_t'], HSS_args['drho'],  HSS_args['outlier'],\
                    HSS_args['pointsize'], HSS_args['mask'], HSS_args['filename'],
                path_plot, HSS_args['verbose'], HSS_args['rho_edge'])
    
    return 

def circular_mask(x, y, cent, radius):
    """
    Create a fake circular grap in a stellar stream
    
    
    Input: center, 3d-radius all in kpc
    
    returns a mask array
    """
    #compute absolute distance 
    r= ((x-cent[0])**2+(y-cent[1])**2)**0.5
    return  r> radius
```

```python
df = ms.FardalStreamDF()
prog_w0 = gd.PhaseSpacePosition(pos=[10., 1, 1.] * u.kpc,
                                vel=[0, 170, 0.] * u.km/u.s)
gen = ms.MockStreamGenerator(df, H)
prog_mass = 5e4 * u.Msun
stream, prog = gen.run(prog_w0, prog_mass,
                       dt=1 * u.Myr, n_steps=1000)
```

```python
#creating mock stream
fig, ax=plt.subplots()
ax.scatter(stream.x, stream.y, s=1, c='k', marker='.')
ax.set(xlabel='x (kpc)', ylabel='y (kpc)')
ax.minorticks_on()
```

```python
stream_r=(np.array([stream.x.value**2+stream.y.value**2+stream.z.value**2])**0.5).flatten()
```

```python
#stream_r.l
```

```python
x, y=stream.x.value/stream_r, stream.y.value/stream_r
mask=~np.logical_or(np.logical_or(x <-0.75, x>-0.5), np.logical_or(y <-2, y>-0.3))
```

```python
#creating mock stream
fig, ax=plt.subplots()
ax.scatter(x[mask], y[mask], s=1, c='k', marker='.')
ax.set(xlabel='x (kpc)', ylabel='y (kpc)')
ax.minorticks_on()
```

```python
x= x[mask][::2]
y=y[mask][::2]
```

```python
#generate uniform background
back_x, back_y= generate_uniform_background(2)
```

```python
#concatenate arrays 
comb_x, comb_y= np.concatenate([x , back_x+x.mean()] ), \
np.concatenate([y, back_y+y.mean()] )
```

```python
#
```

```python
#creating mock stream
fig, ax=plt.subplots()
ax.scatter(comb_x, comb_y, s=1, c='k', marker='.')
ax.set(xlabel='x ', ylabel='y')
ax.minorticks_on()
```

```python
#set HSS parameters
HSS_args={'unit':"unitless",
          'd_galaxy': 1 , #this probably doesn't apply                     
          'delta_t': .1,     
          'drho' : 0.01*3,     
          'outlier': 150, 
          'pointsize': 1,
          'mask': False,
          'verbose': True,
          'rho_edge': False,
         'filename': None}
```

```python
#for dr in [1e-3, 2e-3, 2e-3,  2e-3, 1e-2]:
#dr=0.06
dr= 0.01*3
HSS_args['drho']=dr
HSS_args['filename']='mockstream_drho_{}'.format(dr)
plot_hss([comb_x, comb_y],    HSS_args)
```

```python
#read in a real stream 
data_file='andromeda_mhalo5.0e+08_mstream5.0e+04_rhalo20.0 pc_tmax1.0e+03_tmin0.0e+00_vmax-300intergated_to_1.0e+03.npy'
real_stream= np.load(path_data+'/'+data_file, allow_pickle=True).flatten()[0]
real_stream_coord=SkyCoord(x=real_stream['x'], y=real_stream['y'], \
                           z=real_stream['z'], frame='galactocentric')
```

```python
real_stream.keys()
```

```python
#convert to ra and dec or stream's coordinates
rl_x0, rl_y0=real_stream_coord.icrs.ra, real_stream_coord.icrs.dec
```

```python
#circular mask try different centers
mask=circular_mask(rl_x0.value, rl_y0.value, (190,-20), 50)
```

```python
#creating mock stream
fig, ax=plt.subplots(figsize=(12, 4))
ax.scatter(rl_x0, rl_y0, s=1,  marker='.')
ax.scatter(rl_x0[mask], rl_y0[mask], s=1, c='k', marker='.')
ax.set(xlabel='ra (deg)', ylabel='dec (deg)')
ax.minorticks_on()
plt.savefig(path_plot+'/real_stream.pdf')
```

```python
#add background
rl_x, rl_y=rl_x0[mask].value, rl_y0[mask].value
rl_back_x, rl_back_y= generate_uniform_background(0.5)
rl_back_x, rl_back_y=rl_back_x*3.5, rl_back_y*3.5
rl_back_x=rl_back_x*np.ptp(rl_x)
rl_back_y=rl_back_y*np.ptp(rl_y)
#concatenate arrays 
rl_comb_x, rl_comb_y= np.concatenate([rl_x , rl_back_x+rl_x.mean()] ), \
np.concatenate([rl_y, rl_back_y+rl_y.mean()] )
```

```python
#creating mock stream
fig, ax=plt.subplots(figsize=(8, 6))
ax.scatter(rl_comb_x, rl_comb_y, s=1, c='k', marker='.')
ax.set(xlabel='x (unit)', ylabel='y (unit)')
#ax.minorticks_on()
```

```python
HSS_args_real['filename']='mockstream_real_drho_{}'.format(dr)
plot_hss([rl_comb_x, rl_comb_y],HSS_args_real)
```
